```python
import pandas as pd
import os
pwd = os.getcwd()
```

```python
raw_data = pd.read_csv("C:/Users/hrush/Downloads/Uncleaned_DS_jobs.csv/Uncleaned_DS_jobs.csv")
raw_data
```

```python
drop_columns = raw_data.drop(columns=["index","Founded"])
drop_columns
```

```python
drop_columns["salary estimate"] = drop_columns["Salary Estimate"].str[:11]
drop_columns.drop(columns=["Salary Estimate"],inplace=True)
drop_columns
```

```python
new_df = drop_columns
new_df["min_salary"] = new_df["salary estimate"].str[0:5]
new_df["max_salary"] = new_df["salary estimate"].str[6:]
new_df
```

```python
# checking if there are any null values
new_df.isna().sum()
```

```python
new_df1 = new_df.copy()
```

```python
new_df1.drop(columns=["Competitors","Job Description"],inplace=True)
new_df1
```

```python
# removing the rating attached with the company name
new_df2 = new_df1.copy()
new_df2["company_name"] = new_df2["Company Name"].str[:-4]
new_df2
```

```python
new_df2.drop(columns=["Company Name"],inplace=True)
new_df2
```

```python
#replacing the values
new_df3 = new_df2.copy()
new_df3['max_salary'] = new_df3["max_salary"].apply(lambda x : x.replace('$','').replace('K','').replace('-','').replace('(',''))
new_df3
```

```python
new_df3['min_salary'] = new_df3["min_salary"].apply(lambda x : x.replace('$','').replace('K','').replace('-','').replace('(',''))
new_df3
```

```python
# the data is now cleaned and ready to use,visualize.
# let's export the data to excel file,i.e .xlsx
new_df3.to_excel(pwd + "//Cleaned_datascientist_new.xlsx",index=False)
```
